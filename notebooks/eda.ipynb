# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# dataset loader
file_path = '../data/processed/openfoodfacts_filtered.csv'
df = pd.read_csv(file_path)
```

## 1. Dataset Overview

```python
# display of the first 5 rows
df.head(20)
```

```python
# display the dataset information
df.info()
```

## 2. Nutri-Score Distribution

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='nutriscore_grade', data=df, order=['a', 'b', 'c', 'd', 'e'])
plt.title('Nutri-Score Distribution a-b-c-d-e')
plt.xlabel('Nutri-Score')
plt.ylabel('Count')
plt.show()
```
